Check for missing values and outliers within the horsepower column and remove them.

```python
import altair as alt
```

```python
# check for missing values
df.isnull().any()
```

```python
# check for outliers
alt.Chart(df).mark_boxplot().encode(x='hp', tooltip=['hp']).interactive()
```

```python
# remove outlier(s)
filter = df['hp'].values > 264
df_hp_outlier_rem = df[filter]
df_hp_outlier_rem
```
