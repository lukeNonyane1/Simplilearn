```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
plt.rcParams['figure.figsize'] = (12,6) # handles default matplotlib values
```

```python
df = pd.read_csv('driver-data.csv')
df.head()
```

```python
df.info()
```

```python
# use describe to get the descriptive statisics of the data
df.describe()
```

```python
# k-means library
from sklearn.cluster import KMeans
kmeans_model = KMeans(n_clusters=2)
```

```python
df_analyze = df.drop(columns='id')
```

```python
kmeans_model.fit(df_analyze)
```

```python
# determine cluster center vectors
kmeans_model.cluster_centers_
```

```python
print(kmeans_model.labels_)
print(len(kmeans_model.labels_))
```

```python
print(type(kmeans_model.labels_))
unique, counts = np.unique(kmeans_model.labels_, return_counts=True)
print(dict(zip(unique, counts)))
```

```python
df_analyze['cluster'] = kmeans_model.labels_
sns.set_style('whitegrid') # better suited to plot with heavy data elements
sns.lmplot('mean_dist_day', 'mean_over_speed_perc', data=df_analyze, hue='cluster'
          , palette='coolwarm', size=6, aspect=1, fit_reg=False)
```

```python
kmeans_4 = KMeans(n_clusters=4)
```

```python
kmeans_4.fit(df_analyze)
```

```python
print(kmeans_4.cluster_centers_)
unique, counts = np.unique(kmeans_4.labels_, return_counts=True)
kmeans_4.cluster_centers_
print(dict(zip(unique, counts)))
```

```python
df_analyze['cluster'] = kmeans_4.labels_
sns.set_style('whitegrid')
sns.lmplot('mean_dist_day', 'mean_over_speed_perc', data=df_analyze, hue='cluster'
          , palette='coolwarm', size=6, aspect=1, fit_reg=False)
```
