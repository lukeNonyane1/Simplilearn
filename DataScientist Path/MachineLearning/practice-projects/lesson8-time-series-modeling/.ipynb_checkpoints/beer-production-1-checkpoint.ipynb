DESCRIPTION

The Beer production dataset provides a time series data for monthly beer production in Australia, for the period Jan 1956 â€“ Aug 1995.

**Objective:**
- Check for the stationarity of your data using Rolling Statistics and Dickey-Fuller test.
- If stationarity is present, remove it using differencing in Python.

```python
import numpy as np
import pandas as pd
from datetime import datetime as dt
from statsmodels.tsa.stattools import adfuller, acf, pacf
from statsmodels.tsa.arima_model import ARIMA
import math
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('monthly-beer-production-in-austr.csv')
data.head()
```

### EDA

```python
data.info()
```

```python
data['Month'] = data['Month'].apply(lambda x : dt(int(x[:4]), int(x[5:]), 15)) # add 15 to each month value
data = data.set_index('Month') # set month value as index replacing the 0 based index
data.head() # display first 5 rows of data
```

```python
# assign Monthly beer production values in Australia to ts variable
ts = data['Monthly beer production in Australia']
```

```python
# visual stationary check
plt.plot(ts)
```

We can infer there is increasing trend from the results above.

```python
# transform series using log Decomposition (Rolling Statistics)
ts_log = np.log(ts)
```

```python
# transformation functions using Dickey-Fuller test
def test_stationarity(timeseries):
    rolmean = timeseries.rolling(window=52, center=False).mean()
    rolstd = timeseries.rolling(window=52, center=False).std()
    
    orig = plt.plot(timeseries, color='blue', label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label='Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    print('Results of Dickey-Fuller Test:')
    # initialize Dickey-Fuller test variable
    dftest = adfuller(timeseries, autolag='AIC')
    # initialize test statistic with p-value and other required variables
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used',
                                            'Number of Observations Used']) 
    # initialize critical-value variable
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
```

```python
# Rolling Statistics and Dickey-Fuller test
# apply data to function to see results for stationarity
test_stationarity(data['Monthly beer production in Australia'])
```

Test Statistic (-2.282661) is more than critical (-3.444709) value and moving average is not constant over time. Null hypothesis of Dickey-Fuller test cannot be rejected. This shows that the time series is not stationary.

#### Decomposing the data
Seperate a time series into trend, seasonal effects, and remaining variability assumptions.

```python
# plot moving average
plt.plot(ts_log)
```

```python
# initialize moving average variable
movingAverage = ts_log.rolling(window=12).mean()

# initiliaze moving stanard deviation variable
movingSTD = ts_log.rolling(window=12).std()

plt.plot(ts_log)
plt.plot(movingAverage, color='red')
```

#### Removal of Non-stationarity

```python
ts_log_mv_diff = ts_log - movingAverage
ts_log_mv_diff.head(12)

ts_log_mv_diff.dropna(inplace=True)
ts_log_mv_diff.head(10)
```

```python
# check for stationarity by passing time series data into function
test_stationarity(ts_log_mv_diff)
```

The result of Dickey-Fuller test show that Test Statistic (-3.432559) is less than 5% of critical value (-2.867970). We can conclude that we are 95% confident that this series is stationary. We can reject the null hypothesis.
