### Task 3
- Missing value treatment
    - Explain reason chosen for each variable

```python
# missing/null values in each feature of test
missing_values = []
missing_values_overall_count = test_dataset.isnull().sum()
for i in test_dataset.columns:
    if test_dataset[i].isnull().sum() > 1:
        missing_values.append(i)
        print('missing values for {}: {}'.format(i, test_dataset[i].isnull().sum()))
print('number of columns with missing values:',len(missing_values))
```

```python
# dropping BLOCKID; There are no values
test_dataset.shape,train_dataset.shape
test_dataset.drop(columns='BLOCKID', inplace=True)
train_dataset.drop(columns='BLOCKID', inplace=True)
test_dataset.shape,train_dataset.shape
```

```python
# drop records where used_samples=0; These samples are missing data that will provide value during computation.
used_samples_0 = test_dataset.loc[test_dataset['used_samples'] == 0].index
print('test_data shape before dropping used_samples = 0 records:{}'.format(test_dataset.shape))
test_dataset.drop(index=used_samples_0, inplace=True)
print('test_data shape after dropping used_samples = 0 records:{}'.format(test_dataset.shape))
```

```python
missing_values = []
for i in test_dataset.columns:
    if test_dataset[i].isnull().sum() > 1:
        missing_values.append(i)
        print('missing values for {}: {}'.format(i, test_dataset[i].isnull().sum()))
print('number of columns with missing values:',len(missing_values))
```
